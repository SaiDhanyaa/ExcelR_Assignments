## **ASSOCIATION RULES FOR THE BOOKS**

```python
# Import libraries 
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
# conda install -c conda-forge mlxtend
```

```python
book = pd.read_csv("/content/book.csv")
book.head()
```

# **Apriori Algorithm with support= 0.1 and confidence=0.7**

```python
frequent_itemsets = apriori(book, min_support=0.1, use_colnames=True)
frequent_itemsets.head()
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

```python
##sorting liftratio in acsending order
rules.sort_values('lift',ascending = False)[0:10]
```

```python
rules[rules.lift>1]
```

```python
from matplotlib import pyplot
rules.plot()
pyplot.show()
```

```python
## thus for # **Apriori Algorithm with support= 0.1 and confidence=0.7**, we have got 100 rules
```
